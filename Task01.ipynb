```python
# Import Libraries
import pandas as pd
import matplotlib.pyplot as plt

# Sample Population Data
data = {
    'Country': ['India', 'China', 'USA', 'Indonesia', 'Pakistan'],
    'Population (Millions)': [1428, 1425, 339, 277, 240]
}

# Create DataFrame
df = pd.DataFrame(data)

# Display Data
print(df)

# Create Bar Chart
plt.figure(figsize=(8,5))
plt.bar(df['Country'], df['Population (Millions)'], color='skyblue')

# Labels and Title
plt.title('Population Distribution by Country')
plt.xlabel('Country')
plt.ylabel('Population (Millions)')

# Show Values on Bars
for i, value in enumerate(df['Population (Millions)']):
    plt.text(i, value + 20, str(value), ha='center')

plt.show()
```
